```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial import Polynomial
```

```python
#least to greatest
p1= Polynomial([3,2,1]) # 3 + 2x + x^2
p2 = Polynomial([1,-3,6]) # 1 - 3x + 6x^2
```

```python
p1*p2

#cannot divide because of rational functions ex) p1/p2
#p1/p2
```

```python
p1.degree() #highest power
```

```python
p1.roots() #roots of the polynomial
```

```python
x1,y1 = p1.linspace(n=101, domain=[0,10]) #requires a range (0-10) and creates 100 points
#takes 100 points and calculates the y values
#now we can plot
x2,y2 = p2.linspace(n=101, domain=[0,10])
```

```python
#making a plot

fig, ax = plt.subplots()
ax.plot(x1,y1, label='p1')
ax.plot(x2,y2, label='p2', linestyle='--')
ax.set_title('Polynomial Multiplication', fontweight='bold')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xticks(np.arange(0,11,1))
ax.grid()
ax.minorticks_on()
ax.fill_between(x2, y2, color='orange', alpha=0.1, label='p2')
ax.legend()
plt.show()
```
