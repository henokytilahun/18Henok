```python
#f(x)=sin(x)
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 0.1)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8, 3)) # emplty figure with 8x3 inches

ax.plot(x, y)
```

```python
x,y= np.arange(0, 10, 0.1), np.arange(0, 10, 0.05)
xs, ys = np.meshgrid(x, y) # creates all possible points of x and y
#wont work for xy but will for xs,ys -> 2d vs 3d
zs = np.sin(xs) * np.sin(ys) # f(x,y) = sin(x) * sin(y)
```

```python
from mpl_toolkits.mplot3d import Axes3D #lets us create 3d plots X,y + z
```

```python
fig = plt.figure(figsize=(5, 5)) #blank 3d canvas
ax = fig.add_subplot(111, projection='3d') #manually create an axis

surf = ax.plot_surface(xs, ys, zs, cmap='RdBu') #plot the surface
#HILLS, high values blue, low valyes red

#how to rotate the plot
ax.view_init(elev=45, azim=45) #elevation and azimuth
ax.view_init(elev=0, azim=0) #From the side
ax.view_init(elev=90, azim=0) #From the top
ax.view_init(elev=30, azim=60) #Tilted and askewed

#resacle the z-axis
ax.set_zlim(-1, 1)
```

```python
fig, ax = plt.subplots(figsize=(4, 4))
#ax.imshow(zs) #looks stretched because xs.shape != ys.shape
mymap = ax.imshow(zs, origin='lower', extent=[0, 10, 0, 10]) #to fix this we need to add extent
cbar = fig.colorbar(mymap, ax=ax, orientation='vertical') #add a colorbar
cbar.ax.set_position([0.8, 0.1, 0.03, 0.8]) #move the colorbar to the left
cbar.set_ticks([-1, 1.1, 0.5]) #set the ticks on the colorbar
```
