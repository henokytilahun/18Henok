```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,11,0.1) # interval = 2 -> change to 0.1 to make it smoother
y= x**0.5
#print(y)

x1 = np.arange(0,11,1)
y1 = x1**0.5


#plot x and y values
fig, ax = plt.subplots(figsize=(5,5)) #missing axis labels, title, grid, dots for data, etc.
ax.plot(x, y)
ax.plot(x1, y1, 'ro') #draws dots instead of line
#ax.scatter(x1, y1, color='red', s=1)

ax.set_xlabel('x', fontsize=10)
ax.set_ylabel('y', fontsize=10)
ax.set_title('Square Root Function', fontsize=10)
ax.grid()
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
```

```python
x = np.arange(0,30,0.1) # interval = 2 -> change to 0.1 to make it smoother
y= x**2
#print(y)

x1 = np.arange(0,30,5)
y1 = x1**2


#plot x and y values
fig, ax = plt.subplots(figsize=(5,5)) #missing axis labels, title, grid, dots for data, etc.
ax.plot(x, y, dashes=[2, 2], color='black') #
#dashes https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
ax.plot(x1, y1, 'ro') #draws dots instead of line
#ax.scatter(x1, y1, color='red', s=1)
ax.set_mouseover(True)
ax.fill_between(x, y, color='blue', alpha=0.1)


ax.set_xlabel('x', fontsize=10)
ax.set_ylabel('y', fontsize=10)
ax.set_title(r'$f(x)=(x)^2$', fontsize=15)
ax.grid()
ax.minorticks_on()
ax.annotate('This is the origin', xy=(0, 0), xytext=(0, 200),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.legend(['y=x^2', 'y=x^2 (some points)'])
#ax.set_xlim(0, 10)
#ax.set_ylim(0, 10)

#fig.savefig('myimage.png', bbox_inches='tight', dpi=500)
```
