```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ti, yi = np.arange(0, 10.1, 0.5), np.arange(0, 10.1, 0.5)

ts, ys = np.meshgrid(ti, yi) #makes grid
yprime = ys - ts # y' = y - t

dy, dt = yprime, 0*yprime + 1 # dy/dt = y - t, dx/dt = 1
mag = np.sqrt(dy**2 + dt**2) # magnitude of the vector (changing size of arrow)
```

```python
ys #checking values
```

```python
dt #checking values
```

```python
plt.quiver(ts, ys, dt/mag, dy/mag) #plotting map of arrows (divisions by mag to make arrows same size)
```

```python
#y'(t) = f(y,t)
def my_ode(y,t):
    return y - t
```

```python
my_ode(1, 0) #helps solve dif equations
```

```python
from scipy.integrate import odeint
```

```python
t = np.arange(0, 10, 0.1)
y0 = odeint(my_ode, 2, t) #solves the dif equation
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.plot(t, y0)
ax.plot(t, t+ np.exp(t) - 1)
```

```python
#y'' = g
def my_ode2(y,t):
    y1, y2 = y
    y1prime = y2 #derivitave of position is velocity
    y2prime = -9.81 #derivitave of velocity is acceleration (negative g)
    return [y1prime, y2prime]

t = np.arange(0, 10, 0.1)
y = odeint(my_ode2, [0,50], t) #drop the ball from 10 feet in air with 0 velocity
```

```python
y.shape
```

```python
#y'' = g
def my_ode3(y,t): #introduces air resistance
    y1, y2 = y
    y1prime = y2 #derivitave of position is velocity
    y2prime = -9.81 -0.05*y2 #derivitave of velocity is acceleration (negative g)
    return [y1prime, y2prime]

t = np.arange(0, 10, 0.1)
y = odeint(my_ode2, [0,50], t) #drop the ball from 10 feet in air with 0 velocity
y_air = odeint(my_ode3, [0,50], t) #air resistance
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.plot(t, y[:,0])
ax.plot(t, y_air[:,0])
```
